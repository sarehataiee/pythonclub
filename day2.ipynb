If no condition before **else** is met, **else** block will be executed.

# The while Loop
Simply do something while condition is true.

```python
i = 1
while i < 6:
    print("before i = ", i)
    i += 1
    print("after i = ", i)
```

Print i as long as i is less than 6. That's what this code did!

If you forget to increase i or increase it outside the while loop, it never ends! This is called an infinite loop.

# Arbitrary Arguments, *args
When the exact number of arguments is not clear, we use a different strategy to handle the situation.

```python
def favorite_singers(*singers):
    print(singers)
    print(singers[0])
    print(singers[1])
    print(singers[2])
favorite_singers("Eminem","50cent","Adele")
```

# Keyword Arguments
Can we give arguments to the function in a wrong order?

```python
def simple_func(name,university):
    print("name : " + name)
    print("university : " + university)
simple_func("Ferdowsi","Mehdi")
```

As you can see order matters. But hopefully there is a way to handle this!

```python
simple_func(university="Ferdowsi",name="Mehdi")
```

Problem is fixed. So if we want to give arguments in a different order we should use parameters to do so.

This is called keyword argument syntax or in short kwargs. Key = Value syntax Another example:

```python
def colors(color2,color3,color1):
    print(color1)
    print(color2)
    print(color3)
colors(color1="red",color2="blue",color3="green")
```

### Error handling:
- The **try** block tests a block of code for errors.
- The **except** block handles the error.
- The **else** block executes code when there is no error.
- The **finally** block executes code, regardless of the result.


```python
try:
    print("code start")
    print(1 / 0)
except:
    print("an error occurs")
else: 
    print("no error")
finally:
    print(f"2+4 equals to {2+4}")
```

```python
for amount in [19999, 30000, 30001]:
    print(f"You have {amount} Tomans.")
    try:
        if amount < 29999:
            raise ValueError("Your money is not enough!")
        elif amount == 30000:
            raise ValueError("Just the right value!")
        else:
            print("Happy shopping!")
    except ValueError as e:
        print(e)
    
    print(10*'--')
```
